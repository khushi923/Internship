```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
playstore_df = pd.read_csv("googleplaystore.csv")
reviews_df = pd.read_csv("googleplaystore_user_reviews.csv")

# Display first few rows
print("First 5 rows of Play Store Data:")
print(playstore_df.head())

print("\nFirst 5 rows of User Reviews Data:")
print(reviews_df.head())

# Check basic info
print("\nPlay Store Data Info:")
print(playstore_df.info())

print("\nUser Reviews Data Info:")
print(reviews_df.info())

# Check missing values
print("\nMissing Values in Play Store Data:")
print(playstore_df.isnull().sum())

print("\nMissing Values in User Reviews Data:")
print(reviews_df.isnull().sum())

# Summary statistics
print("\nSummary Statistics for Play Store Data:")
print(playstore_df.describe())

# Visualizing missing values
plt.figure(figsize=(8, 5))
sns.heatmap(playstore_df.isnull(), cmap='coolwarm', cbar=False)
plt.title("Missing Values in Play Store Data")
plt.show()

# Distribution of Ratings
plt.figure(figsize=(8, 5))
sns.histplot(playstore_df['Rating'].dropna(), bins=20, kde=True, color='blue')
plt.title("Distribution of App Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()

# Most common app categories
plt.figure(figsize=(10, 5))
sns.countplot(y=playstore_df['Category'], order=playstore_df['Category'].value_counts().index, palette='viridis')
plt.title("Most Common App Categories")
plt.show()

# Correlation heatmap
numeric_df = playstore_df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(8, 5))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load datasets
playstore_df = pd.read_csv("googleplaystore.csv")
reviews_df = pd.read_csv("googleplaystore_user_reviews.csv")

# Open a file to store textual output
with open("eda_output.txt", "w", encoding="utf-8") as file:
    # Display first few rows
    file.write("First 5 rows of Play Store Data:\n")
    file.write(str(playstore_df.head()) + "\n\n")

    file.write("First 5 rows of User Reviews Data:\n")
    file.write(str(reviews_df.head()) + "\n\n")

    # Check basic info
    file.write("Play Store Data Info:\n")
    file.write(str(playstore_df.info()) + "\n\n")

    file.write("User Reviews Data Info:\n")
    file.write(str(reviews_df.info()) + "\n\n")

    # Check missing values
    file.write("Missing Values in Play Store Data:\n")
    file.write(str(playstore_df.isnull().sum()) + "\n\n")

    file.write("Missing Values in User Reviews Data:\n")
    file.write(str(reviews_df.isnull().sum()) + "\n\n")

    # Summary statistics
    file.write("Summary Statistics for Play Store Data:\n")
    file.write(str(playstore_df.describe()) + "\n\n")

# Save missing values heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(playstore_df.isnull(), cmap='coolwarm', cbar=False)
plt.title("Missing Values in Play Store Data")
plt.savefig("missing_values_heatmap.png")
plt.close()

# Save distribution of ratings
plt.figure(figsize=(8, 5))
sns.histplot(playstore_df['Rating'].dropna(), bins=20, kde=True, color='blue')
plt.title("Distribution of App Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.savefig("ratings_distribution.png")
plt.close()

# Save most common app categories
plt.figure(figsize=(10, 5))
sns.countplot(y=playstore_df['Category'], order=playstore_df['Category'].value_counts().index, palette='viridis')
plt.title("Most Common App Categories")
plt.savefig("common_categories.png")
plt.close()

# Save correlation heatmap
numeric_df = playstore_df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(8, 5))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.savefig("correlation_heatmap.png")
plt.close()

print("EDA completed! Outputs are saved as 'eda_output.txt' and images.")
```
